```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = sns.load_dataset('titanic')
```

```python
dataset.head()
```

```python
# Finding patterns of data.
# Distribution Plots: Distplot for removine line use kde = False in following code
sns.distplot(x = dataset['age'], bins = 10)
```

```python
# Distribution Plots: Joint Plot
# For Plot 1
sns.jointplot(x = dataset['age'], y = dataset['fare'], kind =
'scatter')
# For Plot 2
sns.jointplot(x = dataset['age'], y = dataset['fare'], kind = 'hex')
```

```python
# Distribution Plots: The Rug Plot
sns.rugplot(dataset['fare'])
```

```python
# Categorical Plots: The Bar Plot
sns.barplot(x='sex', y='age', data=dataset)
```

```python
# Categorical Plots: The Count Plot
sns.countplot(x='sex', data=dataset)
```

```python
# Categorical Plots: The Box Plot
sns.boxplot(x='sex', y='age', data=dataset)
```

```python
# more fancy box plot by passing the survived as value to the hue parameter
sns.boxplot(x='sex', y='age', data=dataset, hue="survived")
```

```python
# Categorical Plots: The Violin Plot
sns.violinplot(x='sex', y='age', data=dataset)
```

```python
# fancy by adding categorical variable the violin plot using the hue parameter as shown below:
sns.violinplot(x='sex', y='age', data=dataset, hue='survived')
```

```python
# Advanced Plots: The Strip Plot
sns.stripplot(x='sex', y='age', data=dataset, jitter=False)
```

```python
# To better comprehend thedata, pass True for the jitter parameter which adds some random noise to the data.
sns.stripplot(x='sex', y='age', data=dataset, jitter=True)
```

```python
# add an additional categorical column to strip plot using hue parameter
sns.stripplot(x='sex', y='age', data=dataset, jitter=True, hue='survived')
```

```python
# Advanced Plots: The Swarm Plot
sns.swarmplot(x='sex', y='age', data=dataset)
```

```python
# add another categorical column to the swarm plot using the hue parameter.
sns.swarmplot(x='sex', y='age', data=dataset, hue='survived')
```
